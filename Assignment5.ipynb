```python
dataset = ["<s> I am Student </s>", "<s> Student I am </s>", "<s> I do not like green oranges with milk </s>"]
```

```python
def unigram(dataset):

  import pandas as pd

  # splitting dataset into list of sentences containing word as element
  words = []
  for sentence in dataset:
    words.extend(sentence.split(" "))

  # saving the count of word corresponding to the word in the dictionary
  word_count = {}
  for word in words:
    if word in word_count:
      word_count[word] += 1
    else:
      word_count[word] = 1

  # number of vocabulary
  vocabulary_count = (len(word_count))

  # probability of the word
  for word in word_count:
    word_count[word] = word_count[word] / vocabulary_count

  #return pandas series
  return pd.Series(word_count)
```

```python
print(unigram(dataset))
```

```python
 def bigram(dataset):

  import pandas as pd

  vocabulary = set()
  words = []
  sentences_dict = {}
  word_count = {}

  # spliiting dataset into sentences dictionary and words list
  for sent_no, sentence in enumerate(dataset):
    sentences_dict[f"sentence{sent_no}"] = sentence.split()
    words.extend(sentence.split(" "))

  # counting the frequency of word
  for word in words:
      if word in word_count:
        word_count[word] += 1
      else:
        word_count[word] = 1
      # using to find the vocabulary(non repetative words)
      vocabulary.add(word)

  # initializing the matrix for bigram
  matrix_col = len(vocabulary)
  matrix_row = len(vocabulary)
  matrix = [[0] * matrix_col for _ in range(matrix_row)]


  sentence_len = 0
  word_no = 0

  # iterating over dataset
  for sentence in sentences_dict:
    # length of each sentence
    sentence_len = len(sentences_dict[sentence])
    # iterating over each word in a sentence
    for word_no, word in enumerate(sentences_dict[sentence]):
      # iterating over each ith jth place in matrix
      for j, word_row in enumerate(vocabulary):
        for i, word_col in enumerate(vocabulary):
          # if word of the sentence equal to the one of the words in the column name(first row) and other condition is to avoid out of index error when sentence ends
          if word == word_row and (word_no + 1) < sentence_len:
            # and the next word(word after previous word) is equal to one of the words in the index(first column) then increment the count
            if sentences_dict[sentence][word_no + 1] == word_col:
              matrix[i][j] += 1

  # iterating over each ith jth place in matrix
  for j, word_row in enumerate(vocabulary):
    for i, word_col in enumerate(vocabulary):
      # dividing the count of the word
      for word in word_count:
        if word == word_row:
          matrix[i][j] /= word_count[word]

  #return pandas dataframe
  return pd.DataFrame(matrix, columns=list(vocabulary), index= list(vocabulary))
```

```python
bigram(dataset)
```

```python
import pandas as pd
from collections import Counter
from typing import List

def trigram(dataset: List[str]) -> pd.DataFrame:
    """
    Build a trigram model based on the given dataset.

    Args:
        dataset: List of sentences as strings.

    Returns:
        pandas DataFrame representing the trigram model.
    """

    vocabulary = set()
    words = []
    sentences_dict = {}
    word_count = Counter()

    # splitting the dataset into sentences and words
    for sent_no, sentence in enumerate(dataset):
        sentences_dict[f"sentence{sent_no}"] = sentence.split()
        words.extend(sentence.split())

    # creating count of word and vocabulary
    word_count.update(words)
    vocabulary.update(words)

    # creating matrix
    num_cols = len(vocabulary)
    num_rows = len(vocabulary) ** 2
    matrix = [[0] * num_cols for _ in range(num_rows)]

    # iterating over the sentences
    for sentence in sentences_dict.values():
        # length of each sentence
        sentence_len = len(sentence)
        # iterating over each word in the sentence
        for word_no, word in enumerate(sentence):
            # for avoiding inde out range error when sentence is finish
            if word_no + 2 < sentence_len:
                # storing next word
                next_word = sentence[word_no + 1]
                next_next_word = sentence[word_no + 2]
                # interating over the matrix
                for i, word_row in enumerate(vocabulary):
                    for j, word_col in enumerate(vocabulary):
                        for k, word_col2 in enumerate(vocabulary):
                            if word == word_row and next_word == word_col and next_next_word == word_col2:
                                # converting into 2d matrix and incremening the count, number of col is vocabulary* vocabulary
                                matrix[i * num_cols + j][k] += 1

    # calculate the probabilities in trigram matrix
    for i, word_row in enumerate(vocabulary):
        for j, word_col in enumerate(vocabulary):
            row_sum = sum(matrix[i * num_cols + j])
            if row_sum > 0:
                matrix[i * num_cols + j] = [count / row_sum for count in matrix[i * num_cols + j]]
    # return transposed dataframe
    df = pd.DataFrame(matrix, columns=list(vocabulary), index=[f"{row}_{col}" for row in vocabulary for col in vocabulary])
    df_transposed = df.transpose()

    return df_transposed
```
