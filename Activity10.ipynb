### What are model evaluation measure, list and explain each of them.



*   Accuracy: Accuracy measure the number of correctly classify instances against total instances. It is not a good measure when instances labels are imbalance.
    
    #### Accuracy = (True Positive + True Negative) / (True Positive + False Positive + True Negative + False Negative)
*   Precision: It is number of instances classify as positive against correct positive. It is useful when finding accurate positive is more important than finding more positive from instances.
    #### Precision = True Positive / (True Positive + False Postive)


*   Recall: It is number of total actual positive against predicted positive. It is useful when finding more positive from actual positive is more important than finding correct positive.
    #### Recall = True Positive / (True Positive + False Negative)
*   F1 Score: It is the harmonic mean between recall and precision. It is importance when imbalance class distribution.
    #### F1 Score = (2 * Recall * Precision) / (Recall + Precision)


*   Mean Square Error: It is metric for regression. It measure the average squared diiference between predicted and actual values. Square is used do error is more penalized.
    #### MSE = 1/n summation(yi - h(xi))^2
*   Mean Absolute Error: It is metric for regression. It measure the average absolute diiference between predicted and actual values.
    #### MAE = 1/n summation|yi - h(xi)|
